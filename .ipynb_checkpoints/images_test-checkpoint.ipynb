```python
import os
import cv2

import pandas as pd
import numpy as np
import matplotlib.image as mpimg
```

```python
%pylab inline
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
```

```python
IMAGE_HEIGHT, IMAGE_WIDTH, IMAGE_CHANNELS = 66, 200, 3
INPUT_SHAPE = (IMAGE_HEIGHT, IMAGE_WIDTH, IMAGE_CHANNELS)
```

```python
def load_image(data_dir, image_file):
    return mpimg.imread(os.path.join(data_dir, image_file.strip()))


def process_image(image):
    # Crop
    image = image[60:-25, :, :]
    # Resize
    image = cv2.resize(image, (IMAGE_WIDTH, IMAGE_HEIGHT), cv2.INTER_AREA)
    # RGB to YUV (As done by Nvidia Model)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2YUV)
    return image
```

```python
def image_augument(data_dir, center, left, right, steering_angle, range_x=100, range_y=10):
    # Random choice from 1-3
    choice = np.random.choice(3)
    if choice == 0:  # Left Image
        image, steering_angle = load_image(data_dir, left), steering_angle + 0.2
    elif choice == 1:  # Right Image
        image, steering_angle = load_image(data_dir, right), steering_angle - 0.2
    else:  # Center Image
        image, steering_angle = load_image(data_dir, center), steering_angle

    # Flip Image
    if np.random.rand() < 0.5:
        image = cv2.flip(image, 1)
        steering_angle = -steering_angle

    # Translate Image
    t_x, t_y = range_x * (np.random.rand() - 0.5), range_y * (np.random.rand() - 0.5)
    steering_angle += t_x * 0.002
    t_m = np.float32([[1, 0, t_x], [0, 1, t_y]])
    height, width = image.shape[:2]
    image = cv2.warpAffine(image, t_m, (width, height))

    # Random Shadow in image
    # Credits - 'https://github.com/naokishibuya'
    x1, y1 = IMAGE_WIDTH * np.random.rand(), 0
    x2, y2 = IMAGE_WIDTH * np.random.rand(), IMAGE_HEIGHT
    xm, ym = np.mgrid[0:IMAGE_HEIGHT, 0:IMAGE_WIDTH]

    mask = np.zeros_like(image[:, :, 1])
    mask[(ym - y1) * (x2 - x1) - (y2 - y1) * (xm - x1) > 0] = 1

    # choose which side should have shadow and adjust saturation
    cond = mask == np.random.randint(2)
    s_ratio = np.random.uniform(low=0.2, high=0.5)

    # adjust Saturation in HLS(Hue, Light, Saturation)
    hls = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
    hls[:, :, 1][cond] = hls[:, :, 1][cond] * s_ratio
    image = cv2.cvtColor(hls, cv2.COLOR_HLS2RGB)

    # Change Brightness
    hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    hsv[:, :, 2] = hsv[:, :, 2] * (1.0 + 0.4 * (np.random.rand() - 0.5))
    image = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)

    return image, steering_angle
```

```python
def load_csv():
    data = pd.read_csv('data/driving_log.csv')
    X = data[['center', 'left', 'right']].values
    y = data['steering'].values

    X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.20)
    return (X_train, X_valid, y_train, y_valid)
```

```python
def show_img(i):
    plt.imshow(i)
    plt.show()
    
def s_im(c, c_new, title='title'):
    f = plt.figure()
    f.add_subplot(1, 2, 1).set_title(title)
    plt.imshow(c_new)
    f.add_subplot(1, 2, 2).set_title('original')
    plt.imshow(c)
    plt.show()
```

```python
f = plt.figure()
f.add_subplot(1,3,1).set_title('Center')
plt.imshow(c)
# plt.show()
f.add_subplot(1,3,2).set_title('left')
plt.imshow(l)
# plt.show()
f.add_subplot(1,3,3).set_title('right')
plt.imshow(r)
plt.show()
```

```python
# Flip the image
c_flip = cv2.flip(c, 1)
# show_img(c_flip)
s_im(c, c_flip, 'flip')
```

```python
# Translate the image

range_x=100
range_y=10

t_x, t_y = range_x * (np.random.rand() - 0.5), range_y * (np.random.rand() - 0.5)
t_m = np.float32([[1, 0, t_x], [0, 1, t_y]])
height, width = c.shape[:2]
c_trans = cv2.warpAffine(c, t_m, (width, height))

# show_img(c_trans)
s_im(c, c_trans, 'translated')
```

```python
# Random Shadow in Image
x1, y1 = IMAGE_WIDTH * np.random.rand(), 0
x2, y2 = IMAGE_WIDTH * np.random.rand(), IMAGE_HEIGHT
xm, ym = np.mgrid[0:IMAGE_HEIGHT, 0:IMAGE_WIDTH]

mask = np.zeros_like(c[:, :, 1])
mask[(ym - y1) * (x2 - x1) - (y2 - y1) * (xm - x1) > 0] = 1

cond = mask == np.random.randint(2)
s_ratio = np.random.uniform(low=0.2, high=0.5)

hls = cv2.cvtColor(c, cv2.COLOR_RGB2HLS)
hls[:, :, 1][cond] = hls[:, :, 1][cond] * s_ratio
c_bright = cv2.cvtColor(hls, cv2.COLOR_HLS2RGB)

# Change Brightness
hsv = cv2.cvtColor(c_bright, cv2.COLOR_RGB2HSV)
hsv[:, :, 2] = hsv[:, :, 2] * (1.0 + 0.4 * (np.random.rand() - 0.5))
c_bright = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)

# show_img(c_bright)
s_im(c, c_bright, 'brightness alter')
```

```python
# Processing the image

# Crop
c_crop = c[60:-25, :, :]
# show_img(c_crop)
s_im(c, c_crop, 'cropped')
```

```python
# Resize
c_resize = cv2.resize(c_crop, (IMAGE_WIDTH, IMAGE_HEIGHT), cv2.INTER_AREA)
# show_img(c_resize)
s_im(c, c_resize, 'crop and resized')
```

```python
# Color conversion
c_col = cv2.cvtColor(c_resize, cv2.COLOR_RGB2YUV)
# show_img(c_col)
s_im(c, c_col, 'crop, resize and RGB2YUV')
```
